```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set variables
dt1 = 0.05
dt2 = 0.01
dt3 = 0.005


t_max = 10 * np.pi

def dx_func(y):
    dx = y
    return dx

def dy_func(x):
    dy = -x
    return dy

# Initial conditions
t_0 = 0

x_0 = 0
y_0 = 1
a_0 = 0
aa_0 = 1


# Arrays
t1, t2, t3, x1, x2, x3, y1, y2, y3, a1, a2, a3 = [t_0], [t_0], [t_0], [x_0], [x_0], [x_0], [y_0], [y_0], [y_0], [a_0], [a_0], [a_0]
aa1, aa2, aa3 = [aa_0], [aa_0], [aa_0]

# Appending loop for dt1
while abs(t1[-1] - t_max) > dt1 / 2:
    t_new = t1[-1] + dt1
    
    x_new = x1[-1] + dt1 *   dx_func(y1[-1])
    y_new = y1[-1] + dt1 *   dy_func(x1[-1])

    t1.append(t_new)
    
    x1.append(x_new)
    y1.append(y_new)
    
    a_new = np.sin(t_new)
    a1.append(a_new)

    aa_new = np.cos(t_new)
    aa1.append(aa_new)

    
# Appending loop for dt2
while abs(t2[-1] - t_max) > dt2 / 2:
    t_new = t2[-1] + dt2
    
    x_new = x2[-1] + dt2 *   dx_func(y2[-1])
    y_new = y2[-1] + dt2 *   dy_func(x2[-1])

    t2.append(t_new)
    
    x2.append(x_new)
    y2.append(y_new)

    a_new = np.sin(t_new)
    a2.append(a_new)

    aa_new = np.cos(t_new)
    aa2.append(aa_new)
    
# Appending loop for dt3
while abs(t3[-1] - t_max) > dt3 / 2:
    t_new = t3[-1] + dt3
    
    x_new = x3[-1] + dt3 *   dx_func(y3[-1])
    y_new = y3[-1] + dt3 *   dy_func(x3[-1])

    t3.append(t_new)
    
    x3.append(x_new)
    y3.append(y_new)

    a_new = np.sin(t_new)
    a3.append(a_new)

    aa_new = np.cos(t_new)
    aa3.append(aa_new)
'''
# Plot
plt.plot(t1, x1, label='h = 0.05')
plt.plot(t2, x2, label='h = 0.01')
plt.plot(t3, x3, label='h = 0.005')
plt.plot(t1, a1, '--', label='a', color='black')

plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.tight_layout()
plt.show()
'''
```

```python
# differences
x1 = np.array(x1)
x2 = np.array(x2)
x3 = np.array(x3)

y1 = np.array(y1)
y2 = np.array(y2)
y3 = np.array(y3)

a1 = np.array(a1)

diff1 = abs(x1 - a1)
diff2 = abs(x2 - a2)
diff3 = abs(x3 - a3)

diff11 = abs(x1 - aa1)
diff22 = abs(x2 - aa2)
diff33 = abs(x3 - aa3)

'''
# diff plot 
plt.plot(t1, diff1, label='h = 0.05')
plt.plot(t2, diff2, label='h = 0.01')
plt.plot(t3, diff3, label='h = 0.005')

plt.xlabel('t')
plt.ylabel('Difference between Numerical and Exact Solution')
plt.legend()
plt.tight_layout()
plt.show()
'''
```

```python
# Create a figure with two subplots (1 row, 2 columns)
fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

ax[0].plot(t1, x1, label='h = 0.05')
ax[0].plot(t2, x2, label='h = 0.01')
ax[0].plot(t3, x3, label='h = 0.005')
ax[0].plot(t1, a1, '--', label='Exact Solution', color='black')

ax[0].set_xlabel('t', fontsize=14)
ax[0].set_ylabel('x(t)', fontsize=14)
ax[0].legend(fontsize=12)
ax[0].set_title('(a)', fontsize=18)

ax[1].plot(t1, diff1, label='h = 0.05')
ax[1].plot(t2, diff2, label='h = 0.01')
ax[1].plot(t3, diff3, label='h = 0.005')
ax[1].set_title('(b)', fontsize=18)

ax[1].set_xlabel('t', fontsize=14)
ax[1].set_ylabel('Difference between Numerical and Exact Solution', fontsize=14)
ax[1].legend(fontsize=12)

plt.tight_layout()
plt.savefig("E3a.svg", bbox_inches='tight')
plt.show()

# change all font sizes
```
