```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set variables
dt1 = 0.1
dt2 = 0.5
dt3 = 1
t_max = 10
tau = 2



# Initial conditions
t_0 = 0
N_0 = 10
A_0 = 10

# Arrays for different time steps
t1, N1, A1 = [t_0], [N_0], [A_0]
t2, N2, A2 = [t_0], [N_0], [A_0]
t3, N3, A3 = [t_0], [N_0], [A_0]
```

```python
# Appending loop for dt1
while abs(t1[-1] - t_max) > dt1 / 2:
    t_new = t1[-1] + dt1
    N_new = N1[-1] - N1[-1] / tau * dt1
    A_new = A_0 * np.exp(- t_new/tau)
    t1.append(t_new)
    N1.append(N_new)
    A1.append(A_new)

# Appending loop for dt2 (fixing append issue)
while abs(t2[-1] - t_max) > dt2 / 2:
    t_new = t2[-1] + dt2
    N_new = N2[-1] - N2[-1] / tau * dt2
    A_new = A_0 * np.exp(- t_new/tau)
    t2.append(t_new)
    N2.append(N_new)
    A2.append(A_new)

# Appending loop for dt3 (fixing appending issue)
while abs(t3[-1] - t_max) > dt3 / 2:
    t_new = t3[-1] + dt3
    N_new = N3[-1] - N3[-1] / tau * dt3
    A_new = A_0 * np.exp(- t_new/tau)
    t3.append(t_new)
    N3.append(N_new)
    A3.append(A_new)
    
```

```python
# plot 0.1
plt.plot(t1,N1,'-', label='Approx Values')
plt.plot(t1,A1,'-', label='True Values')
plt.xlabel('Time (t)')
plt.ylabel('Number of Nuclei')
plt.legend()
plt.tight_layout()
plt.show()

# marker size 3 if change to dots
```

```python
# plot 0.5
plt.plot(t2,N2,'-', label='Approx Values')
plt.plot(t2,A2,'-', label='True Values')
plt.xlabel('Time (t)')
plt.ylabel('Number of Nuclei')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# plot 1
plt.plot(t3,N3,'-', label='Approx Values')
plt.plot(t3,A3,'-', label='True Values')
plt.xlabel('Time (t)')
plt.ylabel('Number of Nuclei')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# differences
N1 = np.array(N1)
A1 = np.array(A1)

N2 = np.array(N2)
A2 = np.array(A2)

N3 = np.array(N3)
A3 = np.array(A3)

diff1 = abs(N1 - A1)
diff2 = abs(N2 - A2)
diff3 = abs(N3 - A3)
```

```python
#diff plot 1
plt.plot(t1, diff1, '-')
plt.xlabel('Time (t)')
plt.ylabel('Actual & Approx value difference')
plt.tight_layout()
plt.show()
```

```python
#diff plot 2
plt.plot(t2, diff2, '-', color='tab:orange')
plt.xlabel('Time (t)')
plt.ylabel('Actual & Approx value difference')
plt.tight_layout()
plt.show()
```

```python
#diff plot 3
plt.plot(t3, diff3, '-', color='tab:green')
plt.xlabel('Time (t)')
plt.ylabel('Actual & Approx value difference')
plt.tight_layout()
plt.show()
```
