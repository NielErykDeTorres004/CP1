```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set variables
dt1 = 1
dt2 = 0.5
dt3 = 0.1
t_max = 10
tau = 2



# Initial conditions
t_0 = 0
N_0 = 10
A_0 = 10

# Arrays for different time steps
t1, N1, A1 = [t_0], [N_0], [A_0]
t2, N2, A2 = [t_0], [N_0], [A_0]
t3, N3, A3 = [t_0], [N_0], [A_0]
```

```python
# Appending loop for dt1
while abs(t1[-1] - t_max) > dt1 / 2:
    t_new = t1[-1] + dt1
    N_new = N1[-1] - N1[-1] / tau * dt1
    A_new = A_0 * np.exp(- t_new/tau)
    t1.append(t_new)
    N1.append(N_new)
    A1.append(A_new)

# Appending loop for dt2 
while abs(t2[-1] - t_max) > dt2 / 2:
    t_new = t2[-1] + dt2
    N_new = N2[-1] - N2[-1] / tau * dt2
    A_new = A_0 * np.exp(- t_new/tau)
    t2.append(t_new)
    N2.append(N_new)
    A2.append(A_new)

# Appending loop for dt3 
while abs(t3[-1] - t_max) > dt3 / 2:
    t_new = t3[-1] + dt3
    N_new = N3[-1] - N3[-1] / tau * dt3
    A_new = A_0 * np.exp(- t_new/tau)
    t3.append(t_new)
    N3.append(N_new)
    A3.append(A_new)
    
```

```python
# differences
N1 = np.array(N1)
A1 = np.array(A1)

N2 = np.array(N2)
A2 = np.array(A2)

N3 = np.array(N3)
A3 = np.array(A3)

diff1 = abs(N1 - A1)
diff2 = abs(N2 - A2)
diff3 = abs(N3 - A3)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plotting numerical and exact solutions
ax[0].plot(t1, N1, '-', label='Calculated Values, t = 1 s')
ax[0].plot(t2, N2, '-', label='Calculated Values, t = 0.5 s', color='tab:orange')
ax[0].plot(t3, N3, '-', label='Calculated Values, t = 0.1 s', color='tab:green')
ax[0].plot(t1, A1, '--', label='Exact Solution', color='black')

# Setting labels and legend
ax[0].set_xlabel('Time (s)', fontsize=16)  # Custom font size for x-label
ax[0].set_ylabel('Number of Nuclei', fontsize=16)  # Custom font size for y-label
ax[0].legend(fontsize=12)  # Custom font size for legend
ax[0].set_title('(a)', fontsize=18)


# Plotting differences
ax[1].plot(t1, diff1, '-', label='t = 1 s')
ax[1].plot(t2, diff2, '-', label='t = 0.5 s', color='tab:orange')
ax[1].plot(t3, diff3, '-', label='t = 0.1 s', color='tab:green')
ax[1].set_title('(b)', fontsize=18)

# Setting labels and legend for the second subplot
ax[1].set_xlabel('Time (s)', fontsize=16)
ax[1].set_ylabel('Difference between Exact and Calculated Values', fontsize=15)
ax[1].legend(fontsize=12)

# Final adjustments and saving
plt.tight_layout()
plt.savefig("E1.svg", bbox_inches='tight')
plt.show()
```
