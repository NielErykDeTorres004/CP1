```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set variables
dt = 0.05
t_max = 9

def diff_func(t, x):
    diff = t - x*x
    return diff

# Initial conditions
t_0 = 0

x_01 = 3
x_02 = 1
x_03 = 0
x_04 = -0.7
x_05 = -0.75

# Arrays for different time steps
t, x1, x2, x3, x4, x5 = [t_0], [x_01], [x_02], [x_03], [x_04], [x_05]

# Appending loop for dt1
while abs(t[-1] - t_max) > dt / 2:
    t_new = t[-1] + dt
    
    x_new1 = x1[-1] + dt *   diff_func(t[-1], x1[-1])
    x_new2 = x2[-1] + dt *   diff_func(t[-1], x2[-1]) 
    x_new3 = x3[-1] + dt *   diff_func(t[-1], x3[-1]) 
    x_new4 = x4[-1] + dt *   diff_func(t[-1], x4[-1]) 
    x_new5 = x5[-1] + dt *   diff_func(t[-1], x5[-1]) 
    
    t.append(t_new)
    
    x1.append(x_new1)
    x2.append(x_new2)
    x3.append(x_new3)
    x4.append(x_new4)
    x5.append(x_new5)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot
ax[0].plot(t, x1, label='$x_0 = 3$')
ax[0].plot(t, x2, label='$x_0 = 1$')
ax[0].plot(t, x3, label='$x_0 = 0$')
ax[0].plot(t, x4, label='$x_0 = -0.7$')

ax[0].set_xlabel('t')
ax[0].set_ylabel('x(t)')
ax[0].legend(fontsize=12)
ax[0].set_title('(a)', fontsize=18)

ax[1].plot(t, x1, label='$x_0 = 3$')
ax[1].plot(t, x2, label='$x_0 = 1$')
ax[1].plot(t, x3, label='$x_0 = 0$')
ax[1].plot(t, x4, label='$x_0 = -0.7$')
ax[1].plot(t, x5, label='$x_0 = -0.75$')
ax[1].set_title('(b)', fontsize=18)

ax[1].set_xlabel('t')
ax[1].set_ylabel('x(t)')
ax[1].legend(fontsize=12)




plt.tight_layout()
plt.savefig("E2a.svg", bbox_inches='tight')
plt.show()
```
